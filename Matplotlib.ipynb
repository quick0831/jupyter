# Matplotlib

## `import matplotlib.pyplot as plt`

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-5,5.,100)
y=2*x+1
```

```python
plt.plot(x,y)
plt.show()
```

### linecolor

```python
y=(x**2)/3+2*x-1
plt.plot(x,y,color="red")
plt.show()
```

### available linestyle

=> `:` `--` `-.` `-`

```python
plt.plot(x,(x**3)/3-x/3,linestyle=":")
plt.plot(x,(x**3)/5-x/3,linestyle="--")
plt.plot(x,(x**3)/7-x/3,linestyle="-.")
plt.plot(x,(x**3)/15-x/3,linestyle="-")
plt.show()
```

## Axis

```python
x=np.linspace(-5,5.,100)
y=2*x+1
plt.plot(x,y)
plt.show()
```

### Set limit

```python
plt.ylim(-20,20)
plt.plot(x,y)
plt.show()
```

### Set Label

```python
plt.xlabel("X axis")
plt.plot(x,y)
plt.show()
```

### Set ticks

```python
plt.xticks([-3,0,3],["Low","Medium","High"])
plt.plot(x,y)
plt.show()
```

### Edit axis

```python
ax = plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
plt.plot(x,y)
plt.show()
```

```python
ax = plt.gca()
ax.xaxis.set_ticks_position("bottom")
ax.spines["bottom"].set_position(("data", 0))
plt.plot(x,y)
plt.show()
```

## Legend

```python
plt.plot(x,y,label="Hi")
plt.legend(loc="best")
plt.show()
```
